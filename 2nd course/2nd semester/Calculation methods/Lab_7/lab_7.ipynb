```python
def find_vector_B(A, X):
    n = A.shape[0]
    B = np.zeros((n, 1))
    for i in range(n):
        for j in range(n):
            B[i] += A[i, j] * X[j]
    return B
```

```python
B = find_vector_B(A,X)
print("Вектор B: \n" , B)
```

```python
def LU_ROZKLAD(A):
    n = A.shape[0]
    L = np.zeros((n, n))
    U = np.zeros((n, n))

    for i in range(n):
        # Заповнення верхньої трикутної матриці U
        for k in range(i, n):
            U[i, k] = A[i, k] - sum(L[i, j] * U[j, k] for j in range(i))

        # Заповнення нижньої трикутної матриці L
        L[i, i] = 1
        for k in range(i + 1, n):
            L[k, i] = (A[k, i] - sum(L[k, j] * U[j, i] for j in range(i))) / U[i, i]

    return L, U
```

```python
L , U = LU_ROZKLAD(A)
```

```python
print("L = \n",L)
```

```python
print("U = \n",U)
```

```python
def matrix_vector_dobutok(A, X):
    n, m = A.shape
    result = np.zeros((n, 1))
    for i in range(n):
        for j in range(m):
            result[i] += A[i, j] * X[j]

    return result
```

```python
dobutok = matrix_vector_dobutok(A,X)
print("Добуток матриці та вектора: \n" ,dobutok)
```

```python
def eps_axb(A, x, b):
    n = len(b)
    max_residual = 0

    for i in range(n):
        residual_i = abs(sum(A[i][j] * x[j] for j in range(n)) - b[i])
        max_residual = max(max_residual, residual_i)

    return max_residual
```
