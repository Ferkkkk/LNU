```python
from math import cos
```

```python

x0 = 2
xn = 2 + 7
n = int(input("Введіть значення n (кількість вузлів):"))
h = (xn - x0) / n
```

```python
def frange(start,stop, step=1.0):
  while start < stop:
    yield start
    start +=step
```

```python
def function(x):
  y = (1 / 2) * cos(x-2)
  return y
```

```python
X = list()
for x in frange(x0, xn + 1, h):
  result = function(x)
  print(result)
  X.append(result)
```
