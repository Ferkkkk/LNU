```python
from math import cos,sin,fabs
```

```python
x0 = 0.27718716808958077
h = pow(10,-3)
```

```python
def f(x):
  return 1/2 * cos(x+2)
```

```python
f(x0)
```

```python
def f_p(x):
  return -1/2 * sin(x+2)
```

```python
f_p(x0)
```

```python
def fn(x,h):
  return (f(x+h)-f(x-h))/(2.0*h)
```

```python
fn(x0,h)
```

```python
def eps1(x,h):
  return fabs(fn(x,h)-f_p(x))
```

```python
eps1(x0,h)
```

```python
fn(x0,h)
```

```python
h_values = [pow(10, -i) for i in range(10)]
min_eps = 1
best_h = None

for h in h_values:
    current_eps = eps1(x0, h)
    if current_eps < min_eps:
        min_eps = current_eps
        best_h = h

print("Найкраще значення h0: %.6e" % best_h)
print("Мінімальна похибка:", min_eps)
```

```python
eps1(x0,best_h)
```
