```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from mpl_toolkits.mplot3d import Axes3D
```

```python
def equations(vars):
    x1, x2 = vars
    eq1 = x1**2 + 2 - 3 - x2
    eq2 = (2-3) * (1 - x1) - x2
    return [eq1, eq2]
```

```python
initial_guess = [0, 5]

solution = fsolve(equations, initial_guess)
x1_sol, x2_sol = solution
x1_sol,x2_sol
```

```python
x1_values = np.linspace(-10, 10, 400)
x2_values1 = x1_values**2 + 5
x2_values2 = (2-3) * (1 - x1_values)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x1_values, x2_values1, label=r'$x1^2 + (2-3) - x2 = 0$')
plt.plot(x1_values, x2_values2, label=r'$(2-3)(1-x1) - x2 = 0$')
plt.scatter(x1_sol, x2_sol, color='red', label=f'Solution ({x1_sol:.2f}, {x2_sol:.2f})')
plt.xlabel('$x1$')
plt.ylabel('$x2$')
plt.title('Graphs of the nonlinear equations')
plt.legend()
plt.grid(True)
plt.show()

x1_sol, x2_sol
```

```python
def plot_3d():
    x1 = np.linspace(-10, 10, 400)
    x2 = np.linspace(-10, 10, 400)
    X1, X2 = np.meshgrid(x1, x2)
    Z1 = X1**2 +(2-3) - X2
    Z2 = (2-3) * (1 - X1) - X2

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X1, X2, Z1, alpha=0.5, rstride=100, cstride=100, color='b', edgecolor='none', label=r'$x1^2 + (2-3) - x2 = 0$')
    ax.plot_surface(X1, X2, Z2, alpha=0.5, rstride=100, cstride=100, color='r', edgecolor='none', label=r'$(2-3)(1-x1) - x2 = 0$')
    ax.scatter(0, 5, 0, color='g', s=100, label='Solution (0, 5, 0)')  # Indentation adjusted

    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('Z (Residuals)')
    ax.set_title('3D Graphs of the Nonlinear Equations')
    plt.show()

plot_3d()
```
