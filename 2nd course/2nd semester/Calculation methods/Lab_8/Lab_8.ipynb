```python
n = 100
scale = 2.5
a = 4
b = 8
```

```python
B = np.zeros(n + 1, dtype=np.longdouble)
X0 = np.zeros(n + 1, dtype=np.longdouble)
X1 = np.zeros(n + 1, dtype=np.longdouble)
A = np.zeros((n + 1, n + 1), dtype=np.longdouble)
```

```python
def norma_vector_diff(Y, X, n):
    max_diff = 0
    for i in range(1, n + 1):
        diff = abs(Y[i] - X[i])
        if diff > max_diff:
            max_diff = diff
    return max_diff
```

```python
def norma_matrix(X, n):
    max_sum = 0
    for i in range(1, n + 1):
        row_sum = sum(abs(X[i][j]) for j in range(1, n + 1))
        if row_sum > max_sum:
            max_sum = row_sum
    return max_sum
```

```python
def Method_Zeidelya(A, B, X0, X1, n, k):
    kmax = int(1e5)
    eps = 1e-14
    print("\nМетод Зейделя\n\n")
    k[0] = 1
    while True:
        for i in range(1, n + 1):
            X0[i] = X1[i]
        for i in range(1, n + 1):
            sum = 0.0
            for j in range(1, n + 1):
                if j < i:
                    sum += A[i][j] * X1[j]
                elif j > i:
                    sum += A[i][j] * X0[j]
            X1[i] = (B[i] - sum) / A[i][i]
        k[0] += 1
        if k[0] > kmax or norma_vector_diff(X1, X0, n) < eps:
            break
    if k[0] >= kmax:
        print("Рішення не знайдено!\n")
    else:
        print(f"Рішення знайдено!\n")
        print("Ітерація:", k[0])
```

```python
def Method_Jacobi(A, B, X0, X1, n, k):
    kmax = int(1e5)
    eps = 1e-14
    print("\nМетод Якобі\n\n")
    k[0] = 1
    while True:
        for i in range(1, n + 1):
            X0[i] = X1[i]
        for i in range(1, n + 1):
            sum = 0.0
            for j in range(1, n + 1):
                if j != i:
                    sum += A[i][j] * X0[j]
            X1[i] = (B[i] - sum) / A[i][i]
        k[0] += 1
        if k[0] > kmax or norma_vector_diff(X1, X0, n) < eps:
            break
    if k[0] >= kmax:
        print("Рішення не знайдено!\n")
    else:
        print("Рішення знайдено!\n")
        print("Ітерація:", k[0])
```

```python
def Method_Simple_iteration(A, B, X0, X1, n, k, tau):
    kmax = int(1e5)
    eps = 1e-14
    print("\nМетод простої ітерації\n\n")
    k[0] = 1
    while True:
        for i in range(1, n + 1):
            X0[i] = X1[i]
        for i in range(1, n + 1):
            suma = sum(X0[j] * A[i][j] for j in range(1, n + 1))
            X1[i] = X0[i] - tau * suma + tau * B[i]
        k[0] += 1
        if k[0] > kmax or norma_vector_diff(X1, X0, n) < eps:
            break
    if k[0] >= kmax:
        print("Рішення не знайдено!\n")
    else:
        print("Рішення знайдено!\n")
        print("Ітерація:", k[0])
```

```python
for i in range(1, n + 1):
    suma = 0
    for j in range(1, n + 1):
        A[i][j] = randint(a, b)
        if i != j:
            suma += A[i][j]
    A[i][i] = suma * scale
```
