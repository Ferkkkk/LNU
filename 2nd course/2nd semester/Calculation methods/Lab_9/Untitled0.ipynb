```python
def simple_iteration(f, phi, x0, tol=1e-5, max_iter=100):
  """
  Функція пошуку кореня методом простої ітерації.

  Args:
    f: функція, корінь якої шукається.
    phi: ітераційна функція.
    x0: початкове наближення.
    tol: допустима похибка.
    max_iter: максимальна кількість ітерацій.

  Returns:
    Наближене значення кореня або None, якщо метод не зійшовся.
  """

  for i in range(max_iter):
    xn = phi(x0)
    if abs(f(xn)) < tol:
      return xn
    x0 = xn
  return None

# Приклад використання
def g(x):
  return x**3 - 3*x**2 + 2*x

def phi(x):
  return (3*x**2 - 2*x) / (x**3 - 1)

x = simple_iteration(g, phi, 1.0)
print(f"Наближене значення кореня: {x}")
```
