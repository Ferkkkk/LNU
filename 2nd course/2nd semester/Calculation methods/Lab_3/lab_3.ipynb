```python
from math import cos , factorial , fabs
```

```python
N=20
x0=-2
xn=3
xl=[]
yl=[]
```

```python
h = (xn - x0)/N
h
```

```python
def f(x):
    return x*cos(x+2)
f(0.01)
```

```python
for i in range(N):
    x = x0 + (i * h)
    y = f(x)
    xl.append(x)
    yl.append(y)
```

```python
yl
```

```python
def fact(k):
  if k==0:
    return 1
  else:
    return k*fact(k-1)
```

```python
def Cnk(n,k):
  return fact(n)/ (fact(k)*fact((n-k)))
```

```python
def step(n):
  if n%2:
    return -1
  else:
    return 1
```

```python
def deltaf(n):
  r=0
  for i in range(0,n):
    r+= yl[i]*step(n-i)*Cnk(n,i)
  return r
```

```python
deltaf(10)
```

```python
def fact_mn(t,k):
  mn=1
  if k==0:
    mn=1
  else:
    for i in range(0,k):
      mn*=(t-i)
  return mn
```

```python
fact_mn(5,5)
```

```python
def fact_appr(n,t):
  res=0
  for i in range(0,n):
    res += deltaf(i) * fact_mn(t,i) / fact(i)
  return res
```

```python
fact_appr(20,0.01)
```

```python
def eps(ft,appr):
  return fabs(ft-appr)
```

```python
t=0
n=20
h=0.01
ht=20/(20*n)
start_result=[]
result_appr=[]
result_eps=[]
for i in range(0,n):
  start_result.append(f(t))
  result_appr.append(fact_appr(n,t))
  result_eps.append(eps(ft=f(x0+t*h),appr=fact_appr(n,t)))
  t=t+0.01
```
