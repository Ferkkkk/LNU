# Classification

```python
from sklearn import datasets
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold, cross_val_score,  cross_val_predict
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score, \
log_loss, recall_score, precision_score
from sklearn.model_selection import GridSearchCV
sns.set_style("whitegrid")
%matplotlib inline
```

```python
df=pd.read_csv('data/data_iris.csv')
df.head()
```

```python
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']
```

```python
X=df[features].values
y=df.target.values
X_train, X_test, y_train, y_test = train_test_split( X,y, test_size=0.3, random_state=11)
```

```python
clf = RandomForestClassifier(n_estimators=100)
_=clf.fit(X_train,y_train)
```

```python
pred=clf.predict(X_test)
```

```python
confusion_matrix(y_test,pred)
```

```python
classf_report=classification_report(y_test, pred)
print(classf_report)
```

```python
pred_proba=clf.predict_proba(X_test)
```

```python
pred=np.zeros(X.shape[0])
pred
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=10)
for i, (train_index, test_index) in enumerate(kf.split(X)):
    clf = RandomForestClassifier(n_estimators=100)
    print (f'\nFold {i}')
    clf.fit(X[train_index,:], y[train_index])
    pred[test_index]= clf.predict(X[test_index,:])
```

```python
print(classification_report(y, pred))
```

```python
confusion_matrix(y,pred)
```

```python
kf = KFold(n_splits = 5, shuffle = True,  random_state = 10)
clf = RandomForestClassifier(n_estimators=100)
pred = cross_val_predict(clf, X, y, cv = kf, method = 'predict')
df['pred']=pred
```

```python
clf = GaussianNB()
_=clf.fit(X_train,y_train)
pred=clf.predict(X_test)
classf_report=classification_report(y_test, pred)
print(classf_report)
```

```python
clf = svm.SVC()
_=clf.fit(X_train,y_train)
pred=clf.predict(X_test)
classf_report=classification_report(y_test, pred)
print(classf_report)
```

```python
kf = KFold(n_splits = 5, shuffle = True,  random_state = 10)
clf = svm.SVC()
pred = cross_val_predict(clf, X, y, cv = kf, method = 'predict')
classf_report=classification_report(df['target'].values, pred)
print(classf_report)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
_=clf.fit(X_train,y_train)
pred=clf.predict(X_test)
classf_report=classification_report(y_test, pred)
print(classf_report)
```

```python
parameters = {'n_neighbors':[1,3,5,7]}
knn=KNeighborsClassifier()
clf = GridSearchCV(knn, param_grid=parameters,
                      scoring='accuracy',
                      cv=7)
_=clf.fit(X_train,y_train)
clf.best_params_
```
