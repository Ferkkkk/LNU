```python
colors=['red','green','blue']
df['color']=df["target"].apply(lambda x: colors[x])
df.plot.scatter(x='sepal length (cm)',y='sepal width (cm)', c=df['color'])
plt.show()
df.plot.scatter(x='petal length (cm)', y='petal width (cm)', c=df['color'])
plt.show()
df.plot.scatter(x='sepal length (cm)', y='petal width (cm)', c=df['color'])
plt.show()
```

```python
for i in feature_names:
    sns.boxplot(data=df, x='target', y=i)
    plt.show()
```
