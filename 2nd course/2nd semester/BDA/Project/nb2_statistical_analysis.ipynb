Імпорт бібліотек

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, mannwhitneyu
```

Завантаження датасету

```python
df = pd.read_csv('/content/drive/MyDrive/LNU/2 course/2 semester/BDA/Project/Data/student_addiction_dataset_train.csv')
```

Визначення змінних: Визначаємо цільову змінну (Addiction_Class) та ознаки (всі інші стовпці).

```python
target_variable = 'Addiction_Class'
features = df.columns.drop(target_variable)
```

Хі-квадрат для ознак та цілі: Використовуємо chi2_contingency для перевірки наявності статистично значущого зв'язку між кожною ознакою та цільовою змінною.

```python
# Словник для збереження результатів хі-квадрат тесту
chi2_results = {}

# Проходимо по кожній ознаці
for feature in features:
    # Створення таблиці спряженості
    contingency_table = pd.crosstab(df[feature], df[target_variable])

    # Проведення тесту хі-квадрат
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Збереження p-значення у словнику
    chi2_results[feature] = p
```

Хі-квадрат для пар ознак: Аналогічно перевіряємо зв'язок між парами ознак.

```python
sorted_chi2_results = dict(sorted(chi2_results.items(), key=lambda item: item[1]))
```

Порівнюємо розподіл цільової змінної (Addiction_Class) між двома групами, визначеними значеннями певної ознаки (у прикладі - Experimentation). Це дозволяє оцінити, чи є статистично значуща різниця між групами.

```python
feature_name = 'Experimentation'
group1 = df[df[feature_name] == 'Yes'][target_variable]
group2 = df[df[feature_name] == 'No'][target_variable]
u_statistic, p_value = mannwhitneyu(group1, group2)
print(f"U-тест Манна-Уітні для '{feature_name}' та '{target_variable}':")
print(f"  U-статистика: {u_statistic:.2f}")
print(f"  p-значення: {p_value:.4f}")
```
